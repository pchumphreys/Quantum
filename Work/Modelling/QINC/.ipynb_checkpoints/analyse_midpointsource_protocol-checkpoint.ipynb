```python
# Rate vs Fid for BK and Pur (ambitious params)

NoMem = 3
points = np.arange(0.905,0.995,0.01)
k = len(points)
TPur = np.zeros(k); #time of pur in microsec
FPur = np.zeros(k); #output fidelity of pur
xPur = np.zeros(k); #cut-off fidelity of pur
for j,x in enumerate(points):
    print x
    purout = Pur.Pur(NoMem,fid_bound = x,pfc=0.3,pout=0.3,d=50.0, n_iterations=500)
    TPur[j] = purout[0]
    FPur[j] = purout[1]
    xPur[j] = x

TBK = BK.BK(NoMem,pfc=0.3,pout=0.3,d=50.0, n_iterations=1000)[0] #time of in microsec BK
FBK = BK.BK(NoMem,pfc=0.3,pout=0.3,d=50.0, n_iterations=1000)[1] # output fidelity of BK

#Plot
fig, ax = plt.subplots()
ax.plot(FPur,np.array(1e6/TPur), 'bs', label='Pur')
ax.plot(FBK,np.array(1e6/TBK), 'g^', label='BK')


# Now add the legend with some customizations.
legend = ax.legend(loc='upper center', shadow=True)

plt.ylabel('Entanglement generation rate (Hz)')
plt.xlabel('Fidelity')
plt.title('Purification and Barrett and Kok for %d memories' %NoMem)
plt.show()


f = open('/Users/humphreys/Repositories/personal_calcs/Work/Modelling/QINC/Pur_F_t_vs_fid_3mems_contdeadtime.txt','w')
np.savetxt(f,[TPur,FPur])
f.close()

f = open('/Users/humphreys/Repositories/personal_calcs/Work/Modelling/QINC/BK_F_t_vs_fid_3mems_contdeadtime.txt','w')
np.savetxt(f,[TBK,FBK])
f.close()
```

```python
# pur and BK vs distances (ambitious params)

NoMem = 3
nr_points =10
ds = np.linspace(10,100,nr_points)
timesPur =np.zeros(nr_points)
FsPur=np.zeros(nr_points)
timesBK =np.zeros(nr_points)
FsBK=np.zeros(nr_points)

print ds
for i,d in enumerate(ds):
    print 'distance:', d
    purOut = Pur.Pur(NoMem, fid_bound=0.51 ,pfc=0.3,pout=0.3,d=d, n_iterations=500)
    timesBK[i],FsBK[i] = BK.BK(NoMem,pfc=0.3,pout=0.3,d=d, n_iterations=500)
    timesPur[i],FsPur[i] = purOut[0],purOut[1]
print 'finished!'

ratesPur = 1./(timesPur/1.e6)
ratesBK = 1./(timesBK/1.e6)
fig,ax = plt.subplots()
ax.plot(ds,ratesPur,label='rate Pur (s))')
ax.plot(ds,ratesBK,label='rate BK (s))')
ax.legend()
ax.set_xlabel('distance (km)')
ax.set_ylabel('rate (Hz)')
ax.set_title('Purification and BK')

fig,ax2 = plt.subplots()
ax2.plot(ds,FsPur,label='fidelity Pur')
ax2.plot(ds,FsBK,label='fidelity BK')
ax2.set_ylabel('fidelity')
ax2.legend()
ax2.set_xlabel('distance (km)')


plt.savefig('/Users/humphreys/Repositories/personal_calcs/Work/Modelling/QINC/PurBK_F_t_vs_d_3mems.png')
plt.show()
f = open('/Users/humphreys/Repositories/personal_calcs/Work/Modelling/QINC/PurBK_F_t_vs_d_3mems.txt','w')
np.savetxt(f,[ds,FsPur,FsBK,timesPur,timesBK])
f.close()
```

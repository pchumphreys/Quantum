## Brief overview

The NV centre is a point defect in diamond consisting of a substitutional nitrogen - lattice vacancy pair orientated along the [111] crystalline direction (with C$_{3v}$ symmetry): 
<img src="files/NVcentre.png", style="width: 300px;"/> 

It is known to exist in both negative (NV$^−$) and neutral (NV$^0$) charge states. However, quantum information applications focus on the NV$^-$ state due to its favourable qualities. The NV$^0$ state is not suitable as its spin state cannot be read out optically. Although ensemble NV centres have been studied for some time, single NV centres were only first optically isolated in 1997.

### Key characteristics of NV centres for quantum information

1. A bright photostable optical transition that is suitable for the detection of individual centres and single photon generation
2. An optical ZPL fine structure that is dependent on electric, magnetic and strain fields at low temperature, but is approximately electric and strain field independent at room temperature
3. A magnetically resonant and controllable ground state electronic spin that exhibits long coherence times and coupling to proximal electronic and nuclear spins
4. Optical spin-polarisation and readout of the ground state spin
5. Flexibility and robustness in fabrication

## NV centre electronic spin

The $^3A_2$ ground state is a spin triplet, and therefore has three levels: $m_s = -1,0,1$. Within this ground state, the $m_s = \pm 1$ levels are separated by a zero-field splitting of $D\approx 2.88 GHz$ from $m_s = 0$. 

The Hamiltonian for the ground state is given by 
$\mathcal{H}_\mathrm{GS} = D S^2_z + \gamma_e \bf{B}. \bf{S},$
where $\gamma_e = 2.802 MHz/G$ is the gyromagnetic ratio (same as for an isolated electron). Note for idiot quantum photonics guys: 1 Gauss is $10^-4$ Tesla.

Typically we define a two level system, i.e. a qubit, out of the $m_s = 0$ level and either of the $|m_s| = 1$ levels.  This qubit can have coherence times of a few ms at room temperature, and beyond 10 ms at cryogenic temperatures. This is because these electronic levels lie deep within the large (5.5 eV) bandgap of diamond. Second, diamond is 98.9% spinless $^12$C, leading to only slow dephasing from the fluctuating spin bath. Remaining magnetic impurities set the limit on the coherence time.



## Spectra and associated energy levels

The key identifying features of NV$^-$ and NV$^0$ states are the zero phonon lines (ZPLs) at 1.945 eV (637 nm) and 2.156 eV (575 nm), respectively.
<img src="files/spectra.png", style="width:400px;"/> 

Note that the excitation spectra is a function of optical excitation power. This is because photoconversion between NV$^-$ and NV$^0$ states occurs with increasing excitation power, leading to an increase in the NV$^0$ emission relative to NV$^-$. These sharp ZPLs are accompanied by broad vibronic bands that extend over about 100 nm to higher/ lower energy in absorption/emission respectively. For completeness, it should be noted that an additional infrared ZPL at 1.190 eV (1042 nm) exists that is only observable during optical illumination (with energy >1.945 eV), and has been associated with the NV− centre. 

The sharp ZPLs and well defined vibronic bands of NV0 and NV− indicate that the optical transitions occur between discrete defect levels that are deep within the diamond bandgap, such that the continua of valence and conduction band levels are not involved in the optical transitions. 

These spectra result from the energy levels depicted schematically here:
<img src="files/energyLevels.png", style="width:500px;"/> 

It can be seen that the NV$^-$ ZPL results from the transition between the $^3 A_2$ and $^3 E$ triplet states, while the NV$^0$ ZPL occurs due to the transition between the $^2 E$ and $^2 E$ doublet states.

The NV$^−$ centre can also be identified by a zero field magnetic resonance at ∼2.88 GHz. This magnetic resonance occurs between the ms = 0 and ms = $\pm$1 spin sub-levels of the spin triplet ground state $^3A_2$. A further magnetic resonance occurring at ∼1.42 GHz is observed at room temperature and is attributed to a zero field splitting of the spin triplet excited state $^3 E$ that is analogous to that of $^3 A_2$. 

The magnetic resonances of $^3A_2$ and $^3E$ behave as would be expected of triplet spins in a trigonal crystal field and are characterised by the approximately isotropic electron g-factors ∼2.0028 and ∼2.01 respectively (c.f. 2 for a Dirac electron). They exhibit weak interactions with strain and electric fields.

Magnetic resonances can be detected by either conventional electron paramagnetic resonance (EPR) or optically detected magnetic resonance (ODMR) techniques. 


### Low temperature spectra

At low temperatures (<10 K), the excitation spectrum of the NV$^−$ optical ZPL reveals a fine structure that is highly dependent on strain  and electric fields. This implies a fine structure of $^3E$ that is much more complicated than suggested by the associated ODMR at room temperature. 

The low temperature fine structure of $^3E$ and its observed dependence on strain is shown below (analogous results have been obtained using electric fields):
<img src="files/lowTfineStructure.png", style="width:500px;"/> 

It is clear that the presence of strain and/or an electric field splits the $^3E$ fine structure into two branches; the lower branch undergoes two level anti-crossings (as marked in figure 4) that mix the lower fine structure states to a degree dependent on the strain/electric field, whereas the upper branch does not experience any level anti-crossings and the upper fine structure states remain approximately unmixed. The mixing of the lower fine structure states **results in optical spin-flip transitions when the lower branch (E) is excited**. In contrast, the unmixed upper fine structure states maintain optical spin selection rules ($\delta$S = 0, $\delta$ms = 0) and result in optical spin-conserving transitions when the upper branch is excited.

**Optical selection rules:**

|                    | $A_1$      | $A_2$      | $E_1$      | $E_2$      | $E_x$ | $E_y$ |
|--------------------|------------|------------|------------|------------|-------|-------|
| **$^3A_{2,m_s = -1}$** | $\sigma_+$ | $\sigma_+$ | $\sigma_-$ | $\sigma_-$ |       |       |
| **$^3A_{2,m_s = 0}$**  |            |            |            |            | $y$   | $x$   |
| **$^3A_{2,m_s = 1}$**  | $\sigma_-$ | $\sigma_-$ | $\sigma_+$ | $\sigma_+$ |       |   .    |


**However, even the upper branches are not good closed cycle transitions (discussed [here](#ISCrates)), and therefore cannot be used for readout or optical entanglement generation.**  Therefore optical entanglement must proceed by exciting from ms = 0 to $E_0$ levels (usually slightly strain split into $E_x$ and $E_y$ (note the confusing overlap of notations))

The collapse of the low temperature $^3E$ fine structure into a single broad ODMR at room temperature is thought to be due to phonon mediated orbital averaging over the 3E fine structure levels (depicted in Figure 4). This has the consequence that individual centres with different $^3E$ fine structures at low temperature (due to differing crystal strains) will have near identical $^3E$ fine structures at room temperature.

The strain and electric field dependence of the low temperature optical ZPL fine structure enables the transition energy of the ZPL to be tuned and spin- flip/conserving transitions to be selectively excited. 

The approximate strain and electric field independence of the room temperature fine structure alternatively offers some degree of uniformity in the fine structure between centres and simplifies the microwave control of the excited state spin of the centre. 

In addition to their role in orbital averaging, phonons are also believed responsible for the temperature dependent depolarisation and broadening of the optical ZPL. A thorough understanding of the phonon processes in 3E is yet to be obtained.

## Spin dynamics

The spin dynamics of NV− include the transitions that occur between the spin sub-levels of a given triplet state that act to relax any population difference between them, as well as the fluctuating interactions that act to dephase the spin states. 

**These two distinct processes are characterised (as for any spin system) by the spin relaxation rate 1/T1 and the spin homogeneous dephasing rate 1/T2.** 

A further complication arises when measurements are performed on an ensemble of centres, where each centre has different local conditions, or when an ensemble of measurements of a single centre are performed, where each measurement is made in different conditions. For such ensemble measurements, the expectation value of a given observable must be averaged over the relevant distribution of conditions. **This averaging leads to inhomogeneous spin dephasing that is characterised by the rate 1/T2∗, such that T2∗ ≤ T2**. The spin relaxation and dephasing rates are sensitive to various factors including the densities and types of paramagnetic impurities present in the vicinity of the NV centre, the applied magnetic, electric and strain fields as well as electron-phonon interactions and temperature. 

<img src="files/spinDynamics.png", style="width:500px;"/> 

**The ground state spin has the longest room temperature single spin coherence time (T2) of any electronic spin in a solid, being greater than 1.8 ms in highly engineered samples**. The long coherence time of the ground state spin enables its couplings with proximal electronic and nuclear spins in the lattice to be resolved and manipulated.

**Note that these dephasing rates are in the abscence of dynamical decoupling (dd). DD protocols can be used to significantly extend the coherence time, as explored [here](#dynamicalDecoupling)** 

## Spectral diffusion

The spectral diffusion of the NV−optical ZPL is also believed to be related to photoconversion and the ionisation of electron donors in the vicinity of the NV− centre during optical excitation. 

### Optical detection of spin state

The optical detection of the magnetic resonances of the NV− centre is enabled by the differing fluorescence of the ms = 0 and ±1 spin projections. As demonstrated in figure 6, the spin dependence of the NV− fluorescence can be explained by different non-radiative decay pathways for the ms = 0 and ±1 spin projections from $^3E$ to $^3A_2$ via intermediate dark states (believed to be $^1A_1$ and $^1E$). As these non-radiative decay pathways compete with fluorescent decay, the spin projection with the weakest non-radiative decay out of $^3E$ will appear brightest and vice versa. 

<img src="files/opticalDetect.png", style="width:500px;"/> 

**The observed optical enhancement of the $^3A_2$ EPR signal strength implies that the probability of finding the centre in a particular spin projection is polarised by optical excitation, a process known as optical spin-polarisation.**

**The ms = 0 spin projection is the bright projection and also the projection into which the centre is optically polarised.** It is currently believed that the non-radiative decay from the $^3E$ spin sub-levels to the dark states is slower for ms = 0 than for ms = ±1 and that the non-radiative decay out of the dark states to the $^3A_2$ spin sub-levels is similar for both spin sub-levels, thereby leading to a net polarisation into the ms = 0 sub-level. 

Through observing the time-dependence of the centre’s fluorescence (refer to figure 6), it is clear that the process of optical spin-polarisation takes a few optical cycles; the initial fluorescence is representative of the initial spin state of the centre and spin-polarisation is achieved once the optical steady state has been reached. 

Differing from NV−, NV0 does not have detectable magnetic resonances associated with its spin doublet ground (2E) and excited (2A) states.

Regardless of the competing non-radiative processes in the optical cycles of NV− and NV0, both charge states have sufficiently strong fluorescence to be detected as single centres using scanning confocal microscopy. Indeed, both centres have been identified as single photon sources. **The detection of single centres implies certain characteristics of the NV− and NV0 optical cycles: large optical absorption cross-sections, short excited state lifetimes, high quantum yields, and no efficient shelving in a long lived dark state. Furthermore, both charge states have been observed to be extremely photostable under off-resonance excitation (typically 2.32 eV (532 nm)), with no evidence of photobleaching except in extreme condi- tions.**

**However, photoconversion between the charge states and photoblinking under resonant optical excitation has been regularly observed**, but it is not currently understood. The photoconversion efficiency has been observed to be dependent on optical excitation energy and intensity as well as material properties and the method of NV formation.

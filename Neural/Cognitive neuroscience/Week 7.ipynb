```python
import pickle
with open('c10p1.pickle', 'rb') as f:
    data = pickle.load(f)
    
ins = data['c10p1']
```

```python
norm_ins = ins - np.broadcast_to(np.mean(ins,axis=0),(100,2))
plt.scatter(norm_ins[:,0],norm_ins[:,1])
```

```python
w = np.random.rand(2)-0.5
t_steps = 2000
eta = 0.01
alpha = 0
w_s = np.zeros((t_steps,2))
mean_offset = [0,0]
t_ins = norm_ins + np.broadcast_to(mean_offset,(100,2))

for t in range(t_steps):
    u = t_ins[t % len(t_ins)]
    v = w.T @ u
    w = w + eta*(v * u - alpha * v**2 * w)
    w_s[t] = w
    
plt.plot(w_s)
plt.show()
plt.close()
plt.scatter(t_ins[:,0] ,t_ins[:,1])
plt.plot([0,w[0]],[0,w[1]])
```

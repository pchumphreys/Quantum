```python
import pickle

with open('turing_3p4.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
n1,n2,n3,n4 = [data[n] for n in ['neuron1','neuron2','neuron3','neuron4']]
stim = data['stim']

n1_avg,n2_avg,n3_avg,n4_avg = [np.mean(n,axis=0) for n in [n1,n2,n3,n4]]

for n_a in [n1_avg,n2_avg,n3_avg,n4_avg]:
    plt.plot(stim,n_a)
    print(np.max(n_a))
```

```python
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
sample_time = 10

for i,n in enumerate([n1,n2,n3,n4]):
    plt.plot(np.mean(sample_time*n,axis=0), color = colors[i],label=i+1)
    plt.plot(np.var(sample_time*n,axis=0),'--', color = colors[i])
plt.legend()
```
